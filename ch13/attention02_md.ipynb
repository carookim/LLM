### Query (Q): 디코더의 현재 상태 (무엇을 찾고 있는가?)
### Key (K): 인코더의 각 타임스텝 (어디에 정보가 있는가?)
### Value (V): 인코더의 실제 정보 (가져올 내용)
### Attention Score: Q와 K의 유사도
### Attention Weight: Score를 softmax로 정규화
### Context Vector: Weight로 Value를 가중합
