seq2seq
딥러닝 모델의 한 구조적 패턴, 모델의 설계방식 중 하나
```
영문장을 한국어로 번역, 질의 ---> 답변생성, 문장요약 .. 등
자연어 생성과 시퀀스를 시퀀스로 변환하는 기능을 한다.

why?
    입력과 출력의 길이마다 다를수 있는 문제(번역,요약,챗봇)
    고정크기의 피처 벡터는 순서정보와 문맥관계를 충분히 반영 못함
        -> 모든 정보를 하나의 고정 벡터에 담는 구조적 한계가 있기 때문이다.
    단순 분류모델은 시퀀스간 종속적 생성(토큰별 점진적 예측)을 하기에 부적합하다.
```
동작원리
```
1. Encoder : 입력토큰들을 순차적으로 처리 -> 마지막 / 전체 Hidden State 집약 ^
             ^
             입력 시퀀스를 순차적으로 처리하며 각 시점의 hidden state를 생성하고, 
             마지막 hidden state 또는 전체 hidden state들을 이용해 문맥 요약 정보(context)를 만든다.

2. Context : 요약표현
             인코더의 정보를 압축
             인코더가 입력 전체로부터 얻은 요약 표현
             
3. Decoder : Context + 이전에 생성한 토큰을 이용해서 다음 토큰을 반복생성

4. 종료 : 어느 기점까지 위과정을 반복할지 설정할 필요가 있다.
          특수종료토큰 EOS를 사용하여 디코더가 EOS를 생성하면 반복을 중단하도록 한다.
```
