seq2seq
딥러닝 모델의 한 구조적 패턴, 모델의 설계방식 중 하나
```
영문장을 한국어로 번역, 질의 ---> 답변생성, 문장요약 .. 등
자연어 생성과 시퀀스를 시퀀스로 변환하는 기능을 한다.

why?
    입력과 출력의 길이마다 다를수 있는 문제(번역,요약,챗봇)
    고정크기의 피처 벡터는 순서정보와 문맥관계를 충분히 반영 못함
        -> 모든 정보를 하나의 고정 벡터에 담는 구조적 한계가 있기 때문이다.
    단순 분류모델은 시퀀스간 종속적 생성(토큰별 점진적 예측)을 하기에 부적합하다.
```
